# 4. Código de Conducta Específico, Plan de Acción e Indicadores de Seguimiento

Código de Conducta Específico
Este código fue elaborado con base en los riesgos éticos y ambientales detectados en
el proyecto.
El objetivo es establecer lineamientos que promuevan el cumplimiento de normas
ambientales, el uso responsable de los recursos, la correcta gestión de residuos y la
participación de la comunidad educativa en iniciativas de sostenibilidad.
1. Se separarán correctamente los residuos utilizando los contenedores
asignados para cada tipo de desecho.
2. Se utilizarán productos biodegradables o de bajo impacto ambiental en las
labores de limpieza, mantenimiento y operación de las instalaciones.
3. Se participará activamente en las campañas ambientales y se promoverán
bunas prácticas entre compañeros y personal.
4. Será reportado cualquier incumplimiento ambiental o situación que pueda
generar contaminación dentro de la institución.
5. Se cuidarán los recursos institucionales evitando el desperdicio de agua,
energía y materiales de uso común.

Plan de Acción
El cronograma está pensado para un semestre.

| Mes | Actividad                                                                                                                            |
|-----|--------------------------------------------------------------------------------------------------------------------------------------|
| 1   | Detección de áreas con problemas ambientales<br>Instalación o mejora de contenedores para separación de residuos                       |
| 2   | Capacitación a estudiantes, docentes y personal de apoyo sobre manejo de residuos y sostenibilidad                                   |
| 3   | Inicio de campaña de recolección de baterías y residuos reciclables                                                                 |
| 4   | Sustitución gradual de productos de limpieza por alternativas biodegradables                                                        |
| 5   | Auditoría interna para verificar el cumplimiento de las acciones implementadas                                                       |
| 6   | Evaluación de resultados, aplicación de encuestas y presentación de mejoras para el siguiente semestre                               |

Indicadores de Seguimiento
1. Separación adecuada de residuos
* El objetivo es medir la efectividad del programa de separación de
desechos
* Fórmula de cálculo: (Residuos correctamente separados / Total de
residuos generados) × 100
* Unidad de medida: Porcentaje (%)
* Meta esperada: ≥ 80%
* Frecuencia de medición: Mensual
* Su relación con el objetivo estratégico es mejorar la gestión de residuos
y reducir el impacto ambiental

2. Recolección de baterías para reciclaje
* El objetivo es evaluar la participación en la campaña de reciclaje de
baterías
* Fórmula de cálculo: Cantidad total de baterías recolectadas durante el
periodo
* Unidad de medida: Número de baterías
* Meta esperada: 100 baterías por semestre
* Frecuencia de medición: Mensual
* Su relación con el objetivo estratégico es impulsar proyectos de reciclaje
y manejo adecuado de residuos peligrosos

3. Uso de productos biodegradables
* El objetivo es verificar la sustitución de productos contaminantes por
alternativas biodegradables
* Fórmula de cálculo: (Productos biodegradables utilizados / Total de
productos de limpieza utilizados) × 100
* Unidad de medida: Porcentaje (%)
* Meta esperada: ≥ 90%
* Frecuencia de medición: Mensual
* Su relación con el objetivo estratégico es reducir el uso de agentes que
afectan al medio ambiente

4. Participación en capacitaciones ambientales
* El objetivo es medir el involucramiento de la comunidad educativa en
actividades de sensibilización
* Fórmula de cálculo: (Personas capacitadas / Total de personas
convocadas) × 100
* Unidad de medida: Porcentaje (%)

5. Reducción de incidencias ambientales
* El objetivo es evaluar la disminución de prácticas que incumplen normas
ambientales
* Fórmula de cálculo: ((Incidencias iniciales − Incidencias actuales) /
Incidencias iniciales) × 100
* Unidad de medida: Porcentaje (%)
* Meta esperada: Reducción del 50%
* Frecuencia de medición: Trimestral
* Su relación con el objetivo estratégico es garantizar el cumplimiento de
normas ambientales y reducir riesgos éticos

6. Nivel de satisfacción y conciencia ambiental
* El objetivo es conocer la percepción de la comunidad respecto a las
acciones implementadas
* Fórmula de cálculo: Promedio obtenido en encuestas de satisfacción y
conciencia ambiental
* Unidad de medida: Escala de 1 a 5
* Meta esperada: ≥ 4.0
* Frecuencia de medición: Al final del semestre
* Su relación con el objetivo estratégico es evaluar el impacto del proyecto
en la comunidad educativa

# 5. Análisis de rentabilidad de la ética
Cuando se habla sobre rentabilidad, no solamente se debe pensar en beneficios
económicos. Puesto que en el caso de ESFM, implementar acciones enfocadas en el
cuidado del medio ambiente puede generar mejoras en la organización de la escuela,
convivencia de la comunidad y asimismo en la imagen que se proyecta al exterior.
Uno de los beneficios más importantes es el aprovechamiento de los recursos,
acciones más básicas como separar correctamente los residuos, utilizar productos
biodegradables para el mantenimiento de las instalaciones o impulsar programas de
recolección de baterías ayudan a disminuir el impacto ambiental que genera la
actividad diaria dentro de la escuela. Sin embargo, algunos cambios requieren tiempo,
una buena planeación puede evitar desperdicios y favorecer un uso más responsable
de los recursos disponibles.
Por otra parte, el compromiso con el medio ambiente también fortalece la reputación
institucional. En efecto, pues si una escuela promueve prácticas sostenibles, demuestra
interés por los problemas actuales de la sociedad y mantiene una congruencia con la
formación científica y social habla muy bien de ella. Además, este tipo de iniciativas
puede incentivar una mayor participación de estudiantes, profesores y trabajadores.
Otro aspecto que debe considerarse es el clima organizacional. La participación en
actividades ambientales fomenta el trabajo en equipo y desarrolla mayor conciencia
sobre la responsabilidad que cada persona tiene con su entorno. Se ve reflejado
cuando hay una reducción de problemas relacionados con el manejo incorrecto de
residuos y en un mayor interés por mantener espacios limpios y adecuados para todos.
Si se desea conocer si las acciones tienen resultados positivos, se utilizan los
indicadores, un ejemplo de ello es la cantidad de baterías recolectadas para reciclaje,
la correcta clasificación de residuos, etc.

# 6. Conclusiones y compromiso ético
Al aplicar principios éticos relacionados con el cuidado ambiental, pueden aparecer
diferentes obstáculos. En la ESFM, algunos de ellos pueden ser la falta de información
sobre la separación adecuada de residuos, la poca participación de los miembros de
esta comunidad, y las limitaciones de infraestructura que existen para modernizar las
instalaciones.
Sin embargo, estos problemas no significan que sea imposible realizar cambios. La
mejoría de los hábitos ambientales requiere constancia, educación y participación de
toda la comunidad escolar. Pues pequeñas acciones realmente son significativas para
poder generar un cambio en el impacto ambiental.
Nuestra misión como equipo, fue considerar que la ética ambiental debe formar parte
de las actividades cotidianas dentro de la ESFM. Nuestro compromiso consiste en
promover una cultura de respeto hacia el medio ambiente mediante la participación en
campas de reciclaje, separación adecuada y apoyo a iniciativas que permitan reducir el
impacto ambiental de nuestras actividades académicas y personales.
