# Inciso a

Note que en el juego, para ganar se necesita que estando en una casilla anterior o igual a la 20, se tire el dado y de un número igual o mayor a 21. Así mismo, las serpientes y escaleras se les puede poner de manera especial ya que inducen estos cambios:

$$3 \to 11$$

$$15 \to 19$$

$$13 \to 4$$

$$17 \to 10$$

Definimos:

$$E(i)=\text{número de tiradas antes de estar en una posición de al menos 21} i$$

para $i=0,1,\dots,20$.

Así mismo

$$E(i)=0 \quad \text{para todo } i\ge 21$$

pues la partida acabó ya que al estar en una posición mayor a 21 ya ha ganado

La ecuación general de esperanza es:

$$E(i)=1+\frac16\sum E(\text{sucesor después de tirar })$$

donde ese 1 es la esperanza de estar en la $i$-ésima casilla y las demás esperanzas son de estar en dicha casilla dado un tiro. Como el dado es equiprobable de 6 caras, se puede factorizar de la suma, de esto se obtiene un sistema 20x20 de ecuaciones.

## Ecuaciones de cada casilla son:

### Casilla 20

Desde 20 cualquier lanzamiento termina el juego:

$$E(20)=1$$

### Casilla 19

$$E(19)=1+\frac16E(20)$$

### Casilla 18

$$E(18)=1+\frac16(E(19)+E(20))$$

### Casilla 17

$$E(17)=1+\frac16(E(18)+E(19)+E(20))$$

### Casilla 16

$$E(16)=1+\frac16(E(10)+E(18)+E(19)+E(20))$$

### Casilla 15

$$E(15)=1+\frac16(E(16)+E(10)+E(18)+E(19)+E(20))$$

### Casilla 14

$$E(14)=1+\frac16(E(19)+E(16)+E(10)+E(18)+E(19)+E(20))$$

### Casilla 13

$$E(13)=1+\frac16(E(14)+E(19)+E(16)+E(10)+E(18)+E(19))$$

### Casilla 12

$$E(12)=1+\frac16(E(4)+E(14)+E(19)+E(16)+E(10)+E(18))$$

### Casilla 11

$$E(11)=1+\frac16(E(12)+E(4)+E(14)+E(19)+E(16)+E(10))$$

### Casilla 10

$$E(10)=1+\frac16(E(11)+E(12)+E(4)+E(14)+E(19)+E(16))$$

### Casilla 9

$$E(9)=1+\frac16(E(10)+E(11)+E(12)+E(4)+E(14)+E(19))$$

### Casilla 8

$$E(8)=1+\frac16(E(9)+E(10)+E(11)+E(12)+E(4)+E(14))$$

### Casilla 7

$$E(7)=1+\frac16(E(8)+E(9)+E(10)+E(11)+E(12)+E(4))$$

### Casilla 6

$$E(6)=1+\frac16(E(7)+E(8)+E(9)+E(10)+E(11)+E(12))$$

### Casilla 5

$$E(5)=1+\frac16(E(6)+E(7)+E(8)+E(9)+E(10)+E(11))$$

### Casilla 4

$$E(4)=1+\frac16(E(5)+E(6)+E(7)+E(8)+E(9)+E(10))$$

### Casilla 3

$$E(3)=1+\frac16(E(4)+E(5)+E(6)+E(7)+E(8)+E(9))$$

### Casilla 2

$$E(2)=1+\frac16(E(11)+E(4)+E(5)+E(6)+E(7)+E(8))$$

### Casilla 1

$$E(1)=1+\frac16(E(2)+E(11)+E(4)+E(5)+E(6)+E(7))$$

### Casilla 0

$$E(0)=1+\frac16(E(1)+E(2)+E(11)+E(4)+E(5)+E(6))$$

## Forma matricial

El sistema puede escribirse como:

$$A\mathbf{x}=\mathbf{b}$$

donde

$$\mathbf{x}=
\begin{bmatrix}
E(0)\\
E(1)\\
\vdots\\
E(20)
\end{bmatrix}$$

La matriz aumentada se reduce mediante el método de Gauss--Jordan hasta obtener:

$$\left[
\begin{array}{cccc|c}
1&0&0&\cdots&E(0)\\
0&1&0&\cdots&E(1)\\
0&0&1&\cdots&E(2)\\
\vdots&\vdots&\vdots&\ddots&\vdots
\end{array}
\right]$$

Después de aplicar las operaciones elementales por filas se obtiene:

$$E(0)\approx 11.5$$

## Conclusión

El número promedio de tiradas necesarias para terminar el juego es:

$$\boxed{11.5 \text{ tiradas}}$$

```python
import random
import statistics

# =========================
# CONFIGURACIÓN DEL JUEGO
# =========================

FINAL = 20
SIMULACIONES = 100000

# Serpientes y escaleras
especiales = {
    3: 11,   # escalera
    15: 19,  # escalera
    13: 4,   # serpiente
    17: 10   # serpiente
}

# =========================
# FUNCIÓN DEL JUEGO
# =========================

def jugar():

    posicion = 1
    tiros = 0

    while posicion < FINAL:

        dado = random.randint(1, 6)
        tiros += 1

        nueva_posicion = posicion + dado

        # Si se pasa de 20, permanece en la misma casilla
        if nueva_posicion > FINAL:
            nueva_posicion = posicion

        # Aplicar serpientes o escaleras
        if nueva_posicion in especiales:
            nueva_posicion = especiales[nueva_posicion]

        posicion = nueva_posicion

    return tiros

# =========================
# SIMULACIÓN
# =========================

resultados = []

for _ in range(SIMULACIONES):
    pasos = jugar()
    resultados.append(pasos)

# =========================
# RESULTADOS
# =========================

promedio = statistics.mean(resultados)
minimo = min(resultados)
maximo = max(resultados)
desviacion = statistics.stdev(resultados)

print("====================================")
print(" SERPIENTES Y ESCALERAS")
print("====================================")
print(f"Número de simulaciones : {SIMULACIONES}")
print(f"Promedio de tiros      : {promedio:.4f}")
print(f"Mínimo de tiros        : {minimo}")
print(f"Máximo de tiros        : {maximo}")
print(f"Desviación estándar    : {desviacion:.4f}")
print("====================================")
```

# Probabilidad de alcanzar la comida

## Planteamiento

Se tiene el siguiente tablero:

$$ \begin{matrix}
0 & 1 & 7(\text{food}) \\
2 & 3 & 4 \\
8(\text{shock}) & 5 & 6
\end{matrix} $$

Cada vez que el jugador está en una casilla, se mueve aleatoriamente a una de las casillas vecinas conectadas.

La probabilidad de moverse a cada vecino es equiprobable:

$$ \frac{1}{n} $$

donde $n$ es el número de vecinos disponibles.

Queremos calcular la probabilidad de llegar a la comida (casilla 7) antes de llegar al shock (casilla 8), comenzando desde la casilla 0.

Definimos:

$$ P_i = \text{probabilidad de llegar a food antes de shock comenzando en } i $$

Sabemos inmediatamente que:

$$ P_7 = 1 $$

porque si ya estamos en food, el éxito es seguro.

También:

$$ P_8 = 0 $$

porque si llegamos a shock, perdemos.

## Construcción de ecuaciones

### Casilla 0

Desde 0 se puede ir a:

$$ 1 \quad \text{o} \quad 2 $$

cada uno con probabilidad $1/2$.

Entonces:

$$ P_0 = \frac{P_1 + P_2}{2} $$

### Casilla 1

Desde 1 se puede ir a:

$$ 0,\ 3,\ 7 $$

cada uno con probabilidad $1/3$.

Entonces:

$$ P_1 = \frac{P_0 + P_3 + 1}{3} $$

porque $P_7 = 1$.

### Casilla 2

Desde 2 se puede ir a:

$$ 0,\ 3,\ 8 $$

cada uno con probabilidad $1/3$.

Entonces:

$$ P_2 = \frac{P_0 + P_3 + 0}{3} $$

porque $P_8 = 0$.

### Casilla 3

Desde 3 se puede ir a:

$$ 1,\ 2,\ 4,\ 5 $$

cada uno con probabilidad $1/4$.

Entonces:

$$ P_3 = \frac{P_1 + P_2 + P_4 + P_5}{4} $$

### Casilla 4

Desde 4 se puede ir a:

$$ 3,\ 6,\ 7 $$

Entonces:

$$ P_4 = \frac{P_3 + P_6 + 1}{3} $$

### Casilla 5

Desde 5 se puede ir a:

$$ 3,\ 6,\ 8 $$

Entonces:

$$ P_5 = \frac{P_3 + P_6 + 0}{3} $$

### Casilla 6

Desde 6 se puede ir a:

$$ 4 \quad \text{y} \quad 5 $$

Entonces:

$$ P_6 = \frac{P_4 + P_5}{2} $$

## Sistema completo

$$ P_0=\frac{P_1+P_2}{2} $$

$$ P_1=\frac{P_0+P_3+1}{3} $$

$$ P_2=\frac{P_0+P_3}{3} $$

$$ P_3=\frac{P_1+P_2+P_4+P_5}{4} $$

$$ P_4=\frac{P_3+P_6+1}{3} $$

$$ P_5=\frac{P_3+P_6}{3} $$

$$ P_6=\frac{P_4+P_5}{2} $$

Resolviendo el sistema se obtiene:

$$ P_0 = \frac{1}{2} $$

## Resultado

La probabilidad de alcanzar la comida antes del shock comenzando desde la casilla 0 es:

$$ \boxed{\frac{1}{2}} $$

o equivalentemente:

$$ \boxed{50\%} $$

```python
import random

# Grafo de conexiones
vecinos = {
    0: [1, 2],
    1: [0, 3, 7],
    2: [0, 3, 8],
    3: [1, 2, 4, 5],
    4: [3, 6, 7],
    5: [3, 6, 8],
    6: [4, 5]
}

food = 7
shock = 8

def jugar():
    posicion = 0

    while True:
        # Revisar estados terminales
        if posicion == food:
            return True

        if posicion == shock:
            return False

        # Elegir movimiento aleatorio
        posicion = random.choice(vecinos[posicion])

# Simulación
partidas = 1000
victorias = 0

for _ in range(partidas):
    if jugar():
        victorias += 1

porcentaje = (victorias / partidas) * 100

print("Partidas totales:", partidas)
print("Llegaron a food:", victorias)
print("Porcentaje de éxito:", porcentaje, "%")
```
