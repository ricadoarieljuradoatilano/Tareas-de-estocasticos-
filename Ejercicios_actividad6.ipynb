```python
#codigo de chapman
import numpy as np
from math import exp, factorial, sqrt, ceil

R = np.array([
    [0, 2, 3, 0],
    [4, 0, 2, 0],
    [0, 2, 0, 2],
    [1, 0, 3, 0]
], dtype=float)

ri = np.sum(R, axis=1)
r = np.max(ri)

n = len(R)

P_hat = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i == j:
            P_hat[i, j] = 1 - ri[i] / r
        else:
            P_hat[i, j] = R[i, j] / r

print("r =", r)
print("\nP_hat =")
print(P_hat)

def P_t(t):

    M = int(ceil(max(r*t + 5*sqrt(r*t), 20)))

    P = np.zeros((n, n))

    for k in range(M + 1):

        coef = exp(-r*t) * (r*t)**k / factorial(k)

        P += coef * np.linalg.matrix_power(P_hat, k)

    return P, M

for t in [0.5, 1, 5]:

    P, M = P_t(t)

    print("\n-----------------------------------")
    print("t =", t)
    print("M =", M)
    print(np.round(P, 6))

P05, _ = P_t(0.5)
P1, _ = P_t(1)

error = np.max(np.abs(P1 - P05 @ P05))

print("\nError Chapman-Kolmogorov:")
print(error)
```

```python
#uniformizadion codigo
import numpy as np
from math import exp

R = np.array([
    [0, 2, 3, 0],
    [4, 0, 2, 0],
    [0, 2, 0, 2],
    [1, 0, 3, 0]
], dtype=float)

ri = np.sum(R, axis=1)
r = np.max(ri)

n = len(R)

P_hat = np.zeros((n, n))

for i in range(n):
    for j in range(n):

        if i == j:
            P_hat[i, j] = 1 - ri[i] / r
        else:
            P_hat[i, j] = R[i, j] / r

def uniformizacion(t, epsilon=1e-5):

    A = np.eye(n)

    B = exp(-r*t) * np.eye(n)

    c = exp(-r*t)

    suma = c

    k = 1

    while suma < 1 - epsilon:

        c = c * (r*t) / k

        B = B + c*A

        A = A @ P_hat

        suma = suma + c

        k += 1

    M = k - 1

    return B, M

for t in [0.5, 1, 5]:

    P, M = uniformizacion(t)

    print("\n------------------------------------")
    print("t =", t)
    print("M =", M)
    print(np.round(P, 6))
```
