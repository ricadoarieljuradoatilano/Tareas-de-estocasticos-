# Distribución Weibull $We(\alpha,1)$

## Descripción Matemática

La distribución Weibull con parámetros $(\alpha,1)$ tiene:

$$
f(x)=\alpha x^{\alpha-1} e^{-x^\alpha}, \quad x>0
$$

donde:

- $\alpha > 0$ es el parámetro de forma.
- La escala es igual a $1$.

La función de distribución acumulada es:

$$
F(x)=1-e^{-x^\alpha}
$$

La media teórica es:

$$
\mu = \Gamma\left(1+\frac{1}{\alpha}\right)
$$

La varianza teórica es:

$$
\sigma^2=
\Gamma\left(1+\frac{2}{\alpha}\right)
-
\left[
\Gamma\left(1+\frac{1}{\alpha}\right)
\right]^2
$$

La desviación estándar es:

$$
\sigma=\sqrt{\sigma^2}
$$

## Objetivo del Programa

El código en Python realiza:

1. Generación de una muestra aleatoria Weibull.
2. Histograma para verificar la densidad $f(x)$.
3. Gráfica de la muestra ordenada para verificar $F(x)$.
4. Cálculo e impresión de:
    - Media
    - Varianza
    - Desviación estándar

## Código en Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Parametros
alpha = 2
n = 1000

# Generacion de muestra Weibull We(alpha,1)
muestra = np.random.weibull(alpha, n)

# Histograma para verificar f(x)
plt.figure(figsize=(8,5))
plt.hist(muestra, bins=30, density=True,
         alpha=0.6, color='skyblue')

# Densidad teorica
x = np.linspace(0, max(muestra), 200)
fx = alpha * (x**(alpha-1)) * np.exp(-(x**alpha))

plt.plot(x, fx, 'r', lw=2)
plt.title('Histograma y densidad Weibull')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

# Muestra ordenada para verificar F(x)
muestra_ordenada = np.sort(muestra)
F_empirica = np.arange(1, n+1)/n

plt.figure(figsize=(8,5))
plt.plot(muestra_ordenada, F_empirica,
         label='Empirica')

# Funcion teorica
F_teorica = 1 - np.exp(-(x**alpha))

plt.plot(x, F_teorica, 'r', lw=2,
         label='Teorica')

plt.title('Funcion de distribucion acumulada')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid()
plt.show()

# Estadisticos
media = np.mean(muestra)
varianza = np.var(muestra)
desv_std = np.std(muestra)

print("Media:", media)
print("Varianza:", varianza)
print("Desviacion estandar:", desv_std)

````python
import numpy as np
import matplotlib.pyplot as plt


alpha = 2
n = 1000


muestra = np.random.weibull(alpha, n)


plt.figure(figsize=(8,5))

plt.hist(muestra,
         bins=30,
         density=True,
         alpha=0.6,
         color='skyblue',
         label='Histograma')

x = np.linspace(0, max(muestra), 200)

fx = alpha * (x**(alpha-1)) * np.exp(-(x**alpha))

plt.plot(x, fx,
         color='red',
         linewidth=2,
         label='f(x) teórica')

plt.title('Histograma')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()


muestra_ordenada = np.sort(muestra)

F_empirica = np.arange(1, n+1)/n

plt.figure(figsize=(8,5))

plt.plot(muestra_ordenada,
         F_empirica,
         label='F(x) empírica')

F_teorica = 1 - np.exp(-(x**alpha))

plt.plot(x,
         F_teorica,
         color='red',
         linewidth=2,
         label='F(x) teórica')

plt.title('funcion acumulativa')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid()
plt.show()


media = np.mean(muestra)
varianza = np.var(muestra)
desv_std = np.std(muestra)

print("Media =", media)
print("Varianza =", varianza)
print("Desviación estándar =", desv_std)
````

## Distribución Uniforme $U(0,1)$

$$ f(x)=1, \quad 0<x<1 $$

$$ F(x)=x, \quad 0<x<1 $$

La media teórica es:

$$ \mu=\frac{1}{2} $$

La varianza teórica es:

$$ \sigma^2=\frac{1}{12} $$

La desviación estándar es:

$$ \sigma=\sqrt{\frac{1}{12}} $$

## Objetivo del programa

- Generar una muestra aleatoria uniforme.
- Graficar el histograma para verificar $f(x)$.
- Graficar la muestra ordenada para verificar $F(x)$.
- Calcular media, varianza y desviación estándar.

````python
import numpy as np
import matplotlib.pyplot as plt

n = 1000

muestra = np.random.uniform(0,1,n)

plt.figure(figsize=(8,5))

plt.hist(muestra,
         bins=30,
         density=True,
         alpha=0.6,
         color='skyblue',
         label='Histograma')

x = np.linspace(0,1,200)

fx = np.ones_like(x)

plt.plot(x,
         fx,
         color='red',
         linewidth=2,
         label='f(x)')

plt.title('Histograma Uniforme U(0,1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()

muestra_ordenada = np.sort(muestra)

F_empirica = np.arange(1,n+1)/n

plt.figure(figsize=(8,5))

plt.plot(muestra_ordenada,
         F_empirica,
         label='F(x) empirica')

F_teorica = x

plt.plot(x,
         F_teorica,
         color='red',
         linewidth=2,
         label='F(x) teorica')

plt.title('funcionde acumulacion ')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid()
plt.show()

media = np.mean(muestra)
varianza = np.var(muestra)
desv_std = np.std(muestra)

print("Media =", media)
print("Varianza =", varianza)
print("Desviacion estandar =", desv_std)
````

## Distribución Cauchy $C(0,1)$

$$ f(x)=\frac{1}{\pi(1+x^2)}, \quad -\infty < x < \infty $$

$$ F(x)=\frac{1}{\pi}\arctan(x)+\frac{1}{2} $$

La media no existe.

La varianza no existe.

La desviación estándar no existe.

## Objetivo del programa

- Generar una muestra aleatoria Cauchy.
- Graficar el histograma para verificar $f(x)$.
- Graficar la muestra ordenada para verificar $F(x)$.
- Mostrar el comportamiento de la media y varianza muestral.

## Código en Python

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000

muestra = np.random.standard_cauchy(n)

muestra = muestra[np.abs(muestra) < 25]

plt.figure(figsize=(8,5))

plt.hist(muestra,
         bins=40,
         density=True,
         alpha=0.6,
         color='skyblue')

x = np.linspace(-10,10,500)

fx = 1/(np.pi*(1+x**2))

plt.plot(x,fx,'r',linewidth=2)

plt.title('Histograma Cauchy C(0,1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

muestra_ordenada = np.sort(muestra)

F_empirica = np.arange(1,len(muestra_ordenada)+1)/len(muestra_ordenada)

plt.figure(figsize=(8,5))

plt.plot(muestra_ordenada,
         F_empirica)

F_teorica = (1/np.pi)*np.arctan(x) + 0.5

plt.plot(x,
         F_teorica,
         'r',
         linewidth=2)

plt.title('Funcion acumulada Cauchy')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
plt.show()

media = np.mean(muestra)

varianza = np.var(muestra)

desv_std = np.std(muestra)

```

````python
import numpy as np
import matplotlib.pyplot as plt

n = 1000

muestra = np.random.standard_cauchy(n)

muestra = muestra[np.abs(muestra) < 25]

plt.figure(figsize=(8,5))

plt.hist(muestra,
         bins=40,
         density=True,
         alpha=0.6,
         color='skyblue',
         label='Histograma')

x = np.linspace(-10,10,500)

fx = 1/(np.pi*(1+x**2))

plt.plot(x,
         fx,
         color='red',
         linewidth=2,
         label='f(x)')

plt.title('Histograma Cauchy C(0,1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()

muestra_ordenada = np.sort(muestra)

F_empirica = np.arange(1,len(muestra_ordenada)+1)/len(muestra_ordenada)

plt.figure(figsize=(8,5))

plt.plot(muestra_ordenada,
         F_empirica,
         label='F(x) empirica')

F_teorica = (1/np.pi)*np.arctan(x) + 0.5

plt.plot(x,
         F_teorica,
         color='red',
         linewidth=2,
         label='F(x) teorica')

plt.title('funcion acumulada ')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid()
plt.show()

media = np.mean(muestra)
varianza = np.var(muestra)
desv_std = np.std(muestra)

print("Media muestral =", media)
print("Varianza muestral =", varianza)
print("Desviacion estandar muestral =", desv_std)
````
