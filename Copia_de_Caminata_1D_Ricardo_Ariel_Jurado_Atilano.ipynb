## Definicion

Una caminata aleatoria es una sucesion de variables aleatorias
$X_1, X_2, X_3, \dots, X_n$, con probabilidades $p$ y $q$ de cierto resultado,
tales que

$$
p + q = 1.
$$

Asi, por lo visto en clase, cumple con ser un proceso de Markov.
Supongase que para $X_n$ el estado actual es la posicion $i$.
Ahora bien, para su sucesor se tiene la probabilidad $p$ de ubicarse a la derecha
y $q$ a la izquierda.

Esto se admite ya que el suceso $X_{n+1}$ siempre ocurre, es decir,

$$
P(X_{n+1}) = 1.
$$

Por ser un experimento Bernoulli, toma el valor $p$ o $q$, y $0$ en otro caso.

De manera formal:

$$
P(X_{n+1} = i \mid X_n = j) =
\begin{cases}
p & \text{si } i = j + 1 \\
q & \text{si } i = j - 1 \\
0 & \text{en otro caso}
\end{cases}
$$

Cumpliendo asi lo observado en clase, ya que depende unicamente de la posicion anterior.

Por ende, es un proceso de Markov, pues se satisface que:

$$
P(X_{n+1} = x_{n+1} \mid X_1 = x_1, X_2 = x_2, \dots, X_n = x_n)
=
P(X_{n+1} = x_{n+1} \mid X_n = x_n).
$$

Asi entonces, una caminata aleatoria es un modelo que constituye un proceso de Markov dado un experimento Bernoulli.


```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
p = 0.5

#definicion de función, da un paso random con la probabilidad que le das, y #concatena
def tarea(n_pass, p=0.5, x0=0):
    Elpaso = np.random.choice([1, -1], size=n_pass, p=[p, 1-p])
    return np.concatenate(([x0], x0 + np.cumsum(Elpaso)))

#aqui la llamo y grafico en la otra parte desde plot
D1 = tarea(n, p)

mx = np.max(D1)
mm = np.min(D1)
ELCERO = np.any(D1 == 0)

print("alejamiento Máximo:", mx)
print("alejamiento Mínimo:", mm)
print("¿pasa por el cero?:", ELCERO)

plt.figure(figsize=(10,5))
plt.plot(D1)
plt.xlabel("t")
plt.ylabel("pocision")
plt.title(f"caminata de volado")
plt.grid(True)
plt.show()
```
